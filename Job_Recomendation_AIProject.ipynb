# **Preprocessing the Data**

```python

import nltk
nltk.download('punkt')
```

```python
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Convert the text in each column to lowercase
processed_jobs = it.apply(lambda x: x.astype(str).str.lower())

# Remove any punctuation and special characters from the text
processed_jobs = processed_jobs.apply(lambda x: x.str.replace('[^\w\s]',''))

# Tokenize the text by splitting it into individual words
processed_jobs = processed_jobs.apply(lambda x: x.apply(word_tokenize))

processed_jobs
```

```python
import nltk
nltk.download('stopwords')
```

```python
#Remove any stop words
stop_words = set(stopwords.words('english'))
processed_jobs = processed_jobs.apply(lambda x: x.apply(lambda y: [word for word in y if word not in stop_words]))

# Apply stemming 
stemmer = PorterStemmer()
new_data = processed_jobs.apply(lambda x: x.apply(lambda y: [stemmer.stem(word) for word in y]))


# Rejoin the words into a single string
data = new_data.apply(lambda x: x.apply(lambda y: ' '.join(y)))
```

```python
data
```

```python
data.dtypes
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Concatenate the RequiredQual, JobDescription, and Title columns
job_text = data['RequiredQual'] + ' ' + data['JobDescription'] + ' ' + data['Title']

#Combining the RequiredQual and JobDescription into one column
data['Description'] = data['RequiredQual'] + ' ' + data['JobDescription']
data = data.drop(['RequiredQual', 'JobDescription'], axis=1)

# Vectorize the job text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
job_text_vectors = vectorizer.fit_transform(job_text)

print (job_text_vectors)

```

```python
job_text_vectors.shape
```

**Final data to perform clustering**

**Finding optimal number of clusters for Kmeans by Elbow Method**

```python
# Find the optimal number of clusters using the elbow method
inertias = []
max_num_clusters = 20
for i in range(1, max_num_clusters+1):
    km = KMeans(n_clusters=i)
    km.fit(job_text_vectors)
    inertias.append(km.inertia_)

# Plot the elbow curve
plt.plot(range(1, max_num_clusters+1), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

plt.show()
```

**We select 15 clusters from the elbow method, and will assign labels to each clusters now by performing Kmean, also calculate Silhouette Score**

```python
from sklearn.metrics import silhouette_score

# Choose the optimal number of clusters 
num_clusters = 15

# Perform KMeans clustering
km = KMeans(n_clusters=num_clusters, random_state=42)
km.fit(job_text_vectors)

# Calculate silhouette score for KMeans
kmeans_silhouette = silhouette_score(job_text_vectors, km.labels_)
print('KMeans silhouette score:', kmeans_silhouette)
```

```python
# Assign Labels to clusters
data['Kmean Cluster'] = km.labels_
```

**Reducing the TFIDF Vector into 2D and Show Clusters**

```python
from sklearn.manifold import TSNE

# Reduce the dimensionality of the job text vectors to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
job_text_vectors_2d = tsne.fit_transform(job_text_vectors.toarray())

# Create a scatter plot of the data points with the cluster labels as different colors
plt.figure(figsize=(10, 8))
for i in range(num_clusters):
    plt.scatter(job_text_vectors_2d[data['Kmean Cluster'] == i, 0], 
                job_text_vectors_2d[data['Kmean Cluster'] == i, 1],
                s=10, label='Cluster {}'.format(i+1))
plt.legend()
plt.show()
```

**Displaying the Dendogram**

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Create a dendrogram to visualize the clusters at different levels of granularity
Z = linkage(job_text_vectors.toarray(), method='ward')
plt.figure(figsize=(12, 8))
_ = dendrogram(Z, p=30, truncate_mode='level', leaf_font_size=10)
```

**Performing Agglomerative Clustering with Complete Linkage**

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for different numbers of clusters
silhouette_scores = []
for n_clusters in range(2, 21):
    ac = AgglomerativeClustering(n_clusters=n_clusters, linkage='complete')
    cluster_labels_complete = ac.fit_predict(job_text_vectors.toarray())
    silhouette_scores.append(silhouette_score(job_text_vectors.toarray(), cluster_labels_complete))

optimal_n_clusters = np.argmax(silhouette_scores) + 2
print ('Number of optimal clusters: ', optimal_n_clusters,' with score ',max(silhouette_scores)) 

# Plot the silhouette scores for different numbers of clusters
plt.figure(figsize=(8, 8))
plt.plot(range(2, 21), silhouette_scores, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

**Performing Agglomerative Clustering with Single Linkage**

```python
# Calculate the silhouette score for different numbers of clusters
silhouette_scores = []
for n_clusters in range(2, 21):
    ac = AgglomerativeClustering(n_clusters=n_clusters, linkage='single')
    cluster_labels_single = ac.fit_predict(job_text_vectors.toarray())
    silhouette_scores.append(silhouette_score(job_text_vectors.toarray(), cluster_labels_single))

optimal_n_clusters = np.argmax(silhouette_scores) + 2
print ('Number of optimal clusters: ', optimal_n_clusters,' with score ',max(silhouette_scores)) 

# Plot the silhouette scores for different numbers of clusters
plt.figure(figsize=(8, 8))
plt.plot(range(2, 21), silhouette_scores, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

**Performing Agglomerative Clustering with Average Linkage**

```python
# Calculate the silhouette score for different numbers of clusters
silhouette_scores = []
for n_clusters in range(2, 21):
    ac = AgglomerativeClustering(n_clusters=n_clusters, linkage='average')
    cluster_labels_avg = ac.fit_predict(job_text_vectors.toarray())
    silhouette_scores.append(silhouette_score(job_text_vectors.toarray(), cluster_labels_avg))

optimal_n_clusters = np.argmax(silhouette_scores) + 2
print ('Number of optimal clusters: ', optimal_n_clusters,' with score ',max(silhouette_scores)) 

# Plot the silhouette scores for different numbers of clusters
plt.figure(figsize=(8, 8))
plt.plot(range(2, 21), silhouette_scores, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

**Performing Birch Clustering**

```python
from sklearn.cluster import Birch

# Compute silhouette scores for different numbers of clusters
silhouette_scores = []
for n_clusters in range(2, 21):
    birch = Birch(n_clusters=n_clusters)
    cluster_labels = birch.fit_predict(job_text_vectors.toarray())
    silhouette_scores.append(silhouette_score(job_text_vectors.toarray(), cluster_labels))

# Find the optimal number of clusters
optimal_n_clusters = np.argmax(silhouette_scores) + 2
print('Number of optimal clusters:', optimal_n_clusters, 'with score', max(silhouette_scores))

# Plot the silhouette scores for different numbers of clusters
plt.figure(figsize=(8, 8))
plt.plot(range(2, 21), silhouette_scores, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

**Reading the CV text and preprocessing it**

```python
# importing required modules
from PyPDF2 import PdfReader
  
# creating a pdf reader object
reader = PdfReader('CV.pdf')
  
# printing number of pages in pdf file
print(len(reader.pages))
  
# getting a specific page from the pdf file
page = reader.pages[0]
  
# extracting text from page
text = page.extract_text()
print(text)
```

```python
import nltk
nltk.download('wordnet')
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Convert text to lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenize text into words
words = nltk.word_tokenize(text)

# Remove stopwords
stopwords = set(stopwords.words('english'))
words = [word for word in words if word not in stopwords]

# Lemmatize words
lemmatizer = WordNetLemmatizer()
words = [lemmatizer.lemmatize(word) for word in words]

# Join words back into a string
preprocessed_text = ' '.join(words)

print(preprocessed_text)
```

**Now assigning a cluster label to it**

```python
# Convert new sentence to a vector using the same vectorizer object
new_cv_vector = vectorizer.transform([preprocessed_text])

# Predict the cluster for the new sentence
new_cv_cluster = km.predict(new_cv_vector)[0]

print("This CV belongs to cluster:", new_cv_cluster)
```

**For more accuracy, we will calculate the cosine similarity of each tuple in desired cluster with processed text**

```python
from sklearn.metrics.pairwise import cosine_similarity

# Get the clustered data with FinalCluster = new_cv_cluster
cluster_data = ClusteredData[ClusteredData["FinalCluster"] == new_cv_cluster ]

# Vectorize the job text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
cluster_vectors = vectorizer.fit_transform(cluster_data['JobDescription'])

# Vectorize your preprocessed text
preprocessed_text_vector = vectorizer.transform([preprocessed_text])

# Calculate cosine similarities between your preprocessed text vector and each job description vector in cluster 15
similarities = cosine_similarity(preprocessed_text_vector, cluster_vectors)

```

**Showing only top 30 recomendations**

```python
# Get the top 10 tuples with highest cosine similarity
top_30 = similarities.argsort()[0][-30:][::-1]

# Create a new dataframe with the recommended jobs
RecomendedJobs = cluster_data.iloc[top_30][["JobDescription", "Title"]]

# Print the recommended jobs
RecomendedJobs
```
